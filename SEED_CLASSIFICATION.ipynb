```python
#Feature Selection
#Create separate dataframe for all independed varialbes
independent_variables=['area','perimeter','compactness','lengthOfKernel','widthOfKernel','asymmetryCoefficient', 'lengthOfKernelGroove']
X = ssdf[independent_variables]
print(type(X))
print(X)
#Create a Series for the target variable that we want to predict.

#Note: For Classification problems, target variable will be a categorical variab
target_variable='seedType'
y=ssdf[target_variable]
print(type(y))
print(y)
# Convert to numpy array
X_numpy = np.array(X)
y_numpy=np.array(y)
print(type(X_numpy))
print(type(y_numpy))
```

```python
#Data Splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_numpy, y_numpy, test_size=0.2, random_state=1) 
#X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2,random_state=1)


from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
num_trees=3

dt_model = RandomForestClassifier(random_state=1, max_depth= 11, max_samples=100,max_features=5,n_estimators = num_trees)


#Fit RandomForestClassifier model
dt_model.fit(X_train, y_train)

print("Making predictions for the following in train set:")
predicted_values=dt_model.predict(X_train)
true_values=y_train
print("True values are ",true_values)
print("The predicted values are", predicted_values)
#print("Diff", predicted_values-true_values)


training_accuracy=accuracy_score(true_values, predicted_values, normalize=True)
print("Accuracy Score in Training set", training_accuracy)

print("Making predictions for the following in testing set:")
predicted_values=dt_model.predict(X_test)
true_values=y_test
print("True values are ",true_values)
print("The predicted values are", predicted_values)
#print("Diff", predicted_values-true_values)

testing_accuracy=accuracy_score(true_values, predicted_values, normalize=True)
print("Accuracy Score in Testing set", testing_accuracy)
```
